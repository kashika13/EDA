```python
# Cleaning date of journey column
df['Date_of_journey']=df['Date_of_Journey'].str.split('/').str[0]
df['Month_of_journey']=df['Date_of_Journey'].str.split('/').str[1]
df['Year_of_journey']=df['Date_of_Journey'].str.split('/').str[2]

df['Date_of_journey']=df['Date_of_journey'].astype(int)
df['Month_of_journey']=df['Month_of_journey'].astype(int)
df['Year_of_journey']=df['Year_of_journey'].astype(int)
```

```python
df.head()
```

```python
# Deleting date of journey column as it has been splitted into 3 columns
df.drop('Date_of_Journey',inplace=True,axis=1)
df.head()
```

```python
# Cleaning Dep_time column
df['Dep_Hour']=df['Dep_Time'].str.split(':').str[0]
df['Dep_Minute']=df['Dep_Time'].str.split(':').str[1]

df['Dep_Hour']=df['Dep_Hour'].astype(int)
df['Dep_Minute']=df['Dep_Minute'].astype(int)

# Deleting Dep_time column
df.drop('Dep_Time',inplace=True,axis=1)
```

```python
df.head()
```

```python
# Cleaning Arrival time

df['Arrival_Time']=df['Arrival_Time'].apply(lambda x:x.split(' ')[0])
df['Arrival_Hour']=df['Arrival_Time'].str.split(':').str[0]
df['Arrival_Minute']=df['Arrival_Time'].str.split(':').str[1]

df['Arrival_Hour']=df['Arrival_Hour'].astype(int)
df['Arrival_Minute']=df['Arrival_Minute'].astype(int)


df.drop('Arrival_Time',inplace=True,axis=1)
```

```python
## Mapping value

df['Total_Stops']=df['Total_Stops'].map({'non-stop':0, '1 stop':1, '2 stops':2, '3 stops':3, '4 stops':4, np.nan:1})
```

```python
## Cleaning duration


# Function to extract hours and minutes
def extract_duration(duration):
    # Default values
    hours = 0
    minutes = 0
    
    # Check if duration contains hours and extract them
    if 'h' in duration:
        hours = int(duration.split('h')[0].strip())
    
    # Check if duration contains minutes and extract them
    if 'm' in duration:
        minutes_part = duration.split('h')[-1]  # After hours if present, else whole string
        minutes = int(minutes_part.split('m')[0].strip())
    
    return hours, minutes

df['Duration_hour'], df['Duration_minute']=zip(*df['Duration'].apply(extract_duration))


# deleting duration column
df.drop('Duration',axis=1,inplace=True)

```

Using One hot encoding for these columns

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
encoder=OneHotEncoder()
```

```python
encoder.fit_transform(df[['Airline','Source','Destination','Additional_Info']]).toarray()
```

```python
## Converting into dataframe

df_catagorical=pd.DataFrame(encoder.fit_transform(df[['Airline','Source','Destination','Additional_Info']]).toarray(),columns=encoder.get_feature_names_out())
```

```python
## merging dataframes

df.drop(['Airline','Source','Destination','Additional_Info'],axis=1,inplace=True)

df_final=pd.concat([df,df_catagorical],axis=1)
```
